```python
import os
image_path = 'C:\Disco E\Projects\Papers\Generador\TestingData\Fotos Dr. Esteban\Selected'
os.chdir(image_path)
```

```python
import cv2
import math 

dir_list = os.listdir(image_path)
dir_list = dir_list[:-2]
print(dir_list)
tiledResolution = 416

for f in dir_list:
  image = cv2.imread(f)
  totalCropHorizontal = len(image[0])/tiledResolution
  totalCropVertical = len(image)/tiledResolution
  totalCropHorizontal = math.trunc(totalCropHorizontal)
  totalCropVertical = math.trunc(totalCropVertical)

  f_name = f.split(".")

  for i in range(totalCropHorizontal):
    for j in range(totalCropVertical):
      imageOut = image[j*tiledResolution: tiledResolution + j*tiledResolution, i*tiledResolution: tiledResolution + i*tiledResolution] 
      cv2.imwrite(image_path+'/results/'+f_name[0]+"_"+str(i)+"_"+str(j)+'.png',imageOut)
      print(f,"\n")
      cv2.imshow('image',imageOut)

cv2.waitKey(0)
cv2.destroyAllWindows()


```
